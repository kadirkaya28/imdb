```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

# Answer 4.a

```python
class IMDB_Top_Movies():
    def __init__(self, url):
        self.url= url
        self.prepare_movies_data()

    def prepare_movies_data(self):
        # Downloading imdb top 250 movie's data
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html')

        movies = soup.select('td.titleColumn')
        crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]

        ratings = [b.attrs.get('data-value')
                   for b in soup.select('td.posterColumn span[name=ir]')]
        self.data= (movies, crew, ratings)
        
        self.storing_movie_info()


    def storing_movie_info(self):
    # create a empty list for storing movie information
        self.movie_list = []
        movies, crew, ratings = self.data
        # Iterating over movies to extract each movie's details
        for index in range(0, len(movies)):
            # Separating  movie into
            movie_string = movies[index].get_text()
            movie = (' '.join(movie_string.split()).replace('.', ''))
            movie_title = movie[len(str(index))+1:-7]
            year = re.search('\((.*?)\)', movie_string).group(1)
            place = movie[:len(str(index))+1]
            info = {"place": place,
                    "movie_title": movie_title,
                    "year": year,
                    "star_cast": crew[index],
                    "rating": ratings[index][:4]}
            self.movie_list.append(info)

    def show_movies(self, wos= "dataframe"):
        if wos == "dataframe": # woy of show
            df= pd.DataFrame(self.movie_list)
            pd.set_option('display.max_rows', None)
            pd.set_option('display.max_columns', None)
            display(df)
        elif wos == "print":
            for movie in self.movie_list:
                print(movie['place'], 
                      '==>', movie['movie_title'], 
                      '('+movie['year'] + ') ==>', 
                      movie['rating'][:4])
        else:
            raise(ValueError)
```

```python
url = 'http://www.imdb.com/chart/top'
my_imdb_obj= IMDB_Top_Movies(url)
```

```python
class IMDB_Top_Movies():
    def __init__(self, url):
        self.url= url
        self.prepare_movies_data()
        
    def prepare_movies_data(self):
        # Downloading imdb top 250 movie's data
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html')

        movies = soup.select('td.titleColumn')
        crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]

        ratings = [b.attrs.get('data-value')
                   for b in soup.select('td.posterColumn span[name=ir]')]
        self.data= (movies, crew, ratings)
        
        self.storing_movie_info()

    def storing_movie_info(self):
    # create a empty list for storing movie information
        self.movie_list = []
        movies, crew, ratings = self.data
        # Iterating over movies to extract each movie's details
        for index in range(0, len(movies)):
            # Separating  movie into
            movie_string = movies[index].get_text()
            movie = (' '.join(movie_string.split()).replace('.', ''))
            movie_title = movie[len(str(index))+1:-7]
            year = re.search('\((.*?)\)', movie_string).group(1)
            place = movie[:len(str(index))+1]
            info = {"place": place,
                    "movie_title": movie_title,
                    "year": year,
                    "star_cast": crew[index],
                    "rating": float(ratings[index][:3])}
            self.movie_list.append(info)
            
    def get_spec_range(self, param1, param2, wos= "dataframe"):
        params= (param1, param2)
        my_range = filter(lambda m: (m["rating"] >= min(params) and m["rating"] <= max(params)), self.movie_list)
        self.movie_list= list(my_range)
        self.show_movies(wos= wos)
        return pd.DataFrame(my_range)
    
    def show_movies(self, wos= "dataframe"):
        if wos == "dataframe": # woy of show
            df= pd.DataFrame(self.movie_list)
            pd.set_option('display.max_rows', None)
            pd.set_option('display.max_columns', None)
            display(df)
        elif wos == "print":
            for movie in self.movie_list:
                print(movie['place'], 
                      '==>', movie['movie_title'], 
                      '('+movie['year'] + ') ==>', 
                      movie['rating'])
```
